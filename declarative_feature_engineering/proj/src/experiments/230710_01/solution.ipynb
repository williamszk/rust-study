```python
import pandas as pd
```

```python
products = pd.read_csv("./products.csv", sep=";")
products
```

```python
discounts = pd.read_csv("./discounts.csv", sep=";")
discounts
```

```python
clients = pd.read_csv("./clients.csv", sep=";")
clients
```

```python
def create_client_discount(
    clients: pd.DataFrame, discounts: pd.DataFrame
) -> pd.DataFrame:
    clients_df00 = clients["client_id"].to_frame().rename(columns={"id": "client_id"})
    client_discount = clients_df00.merge(discounts, how="cross")
    return client_discount

client_discount_estimated = create_client_discount(clients, discounts)
client_discount_estimated
```

```python
def filter_exclude_client_discount_for_alto_sodium(
    client_discount: pd.DataFrame,
    products: pd.DataFrame,
    clients: pd.DataFrame,
    city: str | list[str],
    canal: str | list[str],
) -> pd.DataFrame:
    df00 = (
        client_discount.merge(products, left_on="product_id", right_on="id")
        .drop(["id", "nome", "category"], axis=1)
        .merge(clients, left_on="client_id", right_on="id")
        .drop(["id", "nome"], axis=1)
    )

    rows_to_exclude = (
        (df00["sodium"] == "ALTO")
        & (df00["canal"].isin(canal))
        & (df00["city"].isin(city))
    )

    df01 = df00[~rows_to_exclude]

    original_columns = list(client_discount.columns)
    df02 = df01.filter(original_columns)

    client_discount_filtered = df02
    return client_discount_filtered


client_discount_filtered = filter_exclude_client_discount_for_alto_sodium(
    client_discount_estimated,
    products,
    clients,
    ["New York"],
    ["UNIVERSITY", "SCHOOL"],
)

client_discount_filtered
```
